```robotframework
%%python module GraphLibrary

from robot.api import logger
from IPython import display

import pandas
import io
import urllib
import base64
import matplotlib.pyplot as plt

from IPython import display

class GraphLibrary:
    def log_as_graph(self, *args):
        """Log list of values as a graph"""
        # Make graph
        graph = pandas.Series(map(int, *args)).plot()
        # Write as image
        imgdata = io.BytesIO()
        graph.figure.savefig(imgdata, format='png')
        imgdata.seek(0)
        plt.clf()
        # Read into HTML
        uri = 'data:image/png;base64,' + urllib.parse.quote(base64.b64encode(imgdata.getbuffer()))
        html = '<img src="' + uri +'"/>'
        # Log into Robot Framework log
        logger.info(html, html=True)
        # Display on the notebook
        return display.HTML(html)
```
